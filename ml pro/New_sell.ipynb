```python
import pandas as pd
```

```python
df = pd.read_csv('Livings.csv')
df.columns=df.columns.str.strip()
```

```python
df=df.dropna()

X = df[['bedroom','sqlt','bathroom']]
y = df[['price']]
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
Xtr=Scaler.fit_transform(X)
Xtr
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=7)
Xtr1=poly.fit_transform(Xtr)

#degree 1 = 0.11
#degree 2 = 0.30
#degree 3 = 0.32
#degree 4 = 0.34
#degree 5 = 0.31
#degree 6 = 0.33
#degree 7 = -3.3
```

```python

from sklearn.linear_model import LinearRegression

lrn = LinearRegression()
lrn.fit(Xtr1,y)
print(lrn.coef_, lrn.intercept_)
```

```python
y_pred = lrn.predict(Xtr1)
df1= df.copy()
df1['pred_price'] = y_pred
df1['pred_price'] = df1['pred_price'].round(2)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mse = mean_squared_error(y,y_pred)
mse
print(mse)

mae = mean_absolute_error(y,y_pred)
mae
print(mae)

r2 = r2_score(y,y_pred)
r2
```
