```python
import pandas as pd
```

```python
df = pd.read_csv('condoPr.csv')

df.columns = df.columns.str.strip()
df = df.dropna()

X = df[['bathroom','Area','bedrooms']]
y =df[['Prices']]
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
#Scaler = MinMaxScaler()
#Xtr=Scaler.fit_transform(X)

scaler = StandardScaler()
Xtr=scaler.fit_transform(X)
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score


#degree 1 = 0.52
#degree 2 = 0.60
#degree 3 = 0.65
#degree 4 = 0.67
#degree 5 = 0.68
#degree 6 = 0.30




```

```python
def ploy_degree_test (Xtr, degree):
    
    poly = PolynomialFeatures(degree=degree)
    Xtr1=poly.fit_transform(Xtr)
    
    lrn = LinearRegression()
    lrn.fit(Xtr1,y)
    
    y_pred = lrn.predict(Xtr1)
    
    
    mse = mean_squared_error(y,y_pred)
    mae = mean_absolute_error(y,y_pred)
    r2 = r2_score(y,y_pred)
    

    return r2,mae,mse , lrn, poly
```

```python
columns = ['degree', 'r2', 'mae', 'mse']
dfresult = pd.DataFrame(columns=columns)
for deg in range(1,20):
     r2, mae, mse, model, polyTr = ploy_degree_test(Xtr, deg)
     
     dfresult.loc[deg] = [deg, r2, mae, mse]  
     
dfresult     
     
```

```python
r2, mae, mse, boon_model, boon_poly = ploy_degree_test(Xtr, 13)
```
